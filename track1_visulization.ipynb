```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import rasterio
from sklearn.preprocessing import StandardScaler
```

```python
#ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
#read all test images
feature_img_folder = "E:/Grss_data/Track1/train/images/"
label_img_folder = "E:/Grss_data/Track1/train/labels/"
```

```python
def read_all_image(img_folder):
#read all images in the folder
    feature_img_list = []
    for filename in os.listdir(img_folder):
        if filename.endswith(".tif"):
            feature_img_list.append(img_folder+filename)
    return feature_img_list
```

```python
feature_img_list = read_all_image(feature_img_folder)
```

```python
#initial a dataframe to save the max and min value of each band of the feature image
feature_0_max_min = pd.DataFrame(columns = ['max_value','min_value'])
feature_1_max_min = pd.DataFrame(columns = ['max_value','min_value'])
feature_2_max_min = pd.DataFrame(columns = ['max_value','min_value'])
feature_3_max_min = pd.DataFrame(columns = ['max_value','min_value'])
feature_4_max_min = pd.DataFrame(columns = ['max_value','min_value'])
feature_5_max_min = pd.DataFrame(columns = ['max_value','min_value'])
#check feature image value
for feature_img_path in feature_img_list:
  feature_img = rasterio.open(feature_img_path)
  feature_img = feature_img.read()
  #save the max and min value of each band of the feature image
  #for i in range(0,5):
    #max
  max_value = np.max(feature_img[0])
  min_value = np.min(feature_img[0])
  #save the max and min value of each band of the feature image
  feature_0_max_min = feature_0_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)
  
  max_value = np.max(feature_img[1])
  min_value = np.min(feature_img[1])
  #save the max and min value of each band of the feature image
  feature_1_max_min = feature_1_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)
  
  max_value = np.max(feature_img[2])
  min_value = np.min(feature_img[2])
  #save the max and min value of each band of the feature image
  feature_2_max_min = feature_2_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)
  
  max_value = np.max(feature_img[3])
  min_value = np.min(feature_img[3])
  #save the max and min value of each band of the feature image
  feature_3_max_min = feature_3_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)
  
  max_value = np.max(feature_img[4])
  min_value = np.min(feature_img[4])
  #save the max and min value of each band of the feature image
  feature_4_max_min = feature_4_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)
  
  max_value = np.max(feature_img[5])
  min_value = np.min(feature_img[5])
  #save the max and min value of each band of the feature image
  feature_5_max_min = feature_5_max_min._append({'max_value':max_value,'min_value':min_value},ignore_index=True)  
  
  
  
  
feature_0_max_min
    
```
